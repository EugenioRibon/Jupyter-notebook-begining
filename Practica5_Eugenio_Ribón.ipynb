fui desarrollando las diferentes iteraciones. Primero traté de escribir una sola palabra en el tablero, después con un bucle puse las demás, creando una variable donde iba almacenando todas las respuestas. Para ello divido las palabras en sus diferentes letras y opero con eso. Finalmente implemento condicionales para el color de las letras en funcion de si estan en la palabra o no y de si estan en la posición correcta o no. Una vez terminado todo esto traté de optimizar el código mediante variables y bucles.


```python
GRIS = "\033[0;47m"
VERDE = "\033[1;42m"
ROJO = "\033[0;41m"
DEFECTO = "\033[0m"
palabra_correcta = "queso".upper()
filas = 0
intentos = 0
palabra_vacia = " " * 5
suma_respuestas = ""

while intentos < 5:
    respuesta = input("Introduce una palabra de 5 letras: ").upper()
    suma_respuestas += respuesta
    filas = 0
    indice1 = 0
    indice2 = 5
    ir1 = 0
    while filas < len(suma_respuestas)//5:
        ir2 = 0
        for letra in suma_respuestas[indice1:indice2]:
            
            if suma_respuestas[ir1] == palabra_correcta[ir2]:
                print(VERDE, letra, DEFECTO, " ", end = "")  
            
            elif suma_respuestas[ir1] in palabra_correcta:
                print(ROJO, letra, DEFECTO, " ", end = "")  
                
            else:
                 print(GRIS, letra, DEFECTO, " ", end = "") 
                    
            ir1 += 1 
            ir2 += 1
        print("\n") 
        filas += 1
        indice1 += 5
        indice2 += 5
        
    while filas < 5:
        for letra in palabra_vacia:
            print(GRIS, letra, DEFECTO, " ", end = "") 
        print("\n") 
        filas += 1
    
    intentos+=1
        
    if respuesta == palabra_correcta:
        print("¡Has acertado, enorabuena!")
        intentos = 5
        
print ("Gracias por jugar, espero que te lo hayas pasado bien. ")
```
